```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/niran/OneDrive/Desktop/Projects/DoorDash Project/final_data.csv")
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the label encoder
label_encoder = LabelEncoder()

# Encode 'store_primary_category'
df['store_primary_category'] = label_encoder.fit_transform(df['store_primary_category'])
```

```python
df.head()
```

```python
# Convert 'created_at' to datetime
df['created_at'] = pd.to_datetime(df['created_at'])

# Extract day of the week (0=Monday, 1=Tuesday, ..., 6=Sunday) and hour of the day
df['day_of_week'] = df['created_at'].dt.dayofweek  # 0-6
df['hour_of_day'] = df['created_at'].dt.hour  # 0-23
```

```python
df.head()
```

```python
df = df.drop(columns=['created_at', 'actual_delivery_time'])
```
