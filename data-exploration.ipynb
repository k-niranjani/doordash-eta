```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
categorical_features = ['market_id', 'order_protocol', 'total_items', 'num_distinct_items']

def plot_categorical_features(df, categorical_features):
    for feature in categorical_features:
        plt.figure(figsize=(4, 4))
        sns.countplot(data=df, x=feature)
        plt.title(f'Bar graph of {feature}')
        plt.xticks(rotation=90)
        plt.show()

# Call the function
plot_categorical_features(df, categorical_features)
```

```python
plt.figure(figsize=(10, 15))
sns.countplot(data=df, y=df['store_primary_category'], order=df['store_primary_category'].value_counts().index)
plt.title(f'Demand for each Cuisine')
plt.xlabel('Count')
plt.ylabel('Cuisine')
plt.show()
```

```python
df['store_id'] = df['store_id'].astype(str)

# Group by store_primary_category and count unique store_id
unique_store_counts = df.groupby('store_primary_category')['store_id'].nunique().sort_values(ascending=False)

# Plotting horizontal bar graph using seaborn
plt.figure(figsize=(10, 15))
sns.barplot(x=unique_store_counts.values, y=unique_store_counts.index)
plt.title('Number of Stores Catering to each Cuisine')
plt.xlabel('Number of Unique Store IDs')
plt.ylabel('Store Primary Category')
plt.show()
```

```python
# Ensure 'created_at' is a datetime object
df['created_at'] = pd.to_datetime(df['created_at'])

# Extract the hour from the 'created_at' column
df['hour'] = df['created_at'].dt.hour

# Group by the hour and calculate the average values
hourly_averages = df.groupby('hour')[['total_onshift_dashers', 'total_busy_dashers', 'total_outstanding_orders']].mean()

max_values = hourly_averages.idxmax()
min_values = hourly_averages.idxmin()
max_averages = hourly_averages.max()
min_averages = hourly_averages.min()

# Combine the max and min values for display
max_min_values = pd.concat([max_values, max_averages, min_values, min_averages], keys=['max_hour', 'max_value', 'min_hour', 'min_value'], axis=1)

print("Hours with maximum average values and their averages:")
print(max_min_values[['max_hour', 'max_value']])
print("\nHours with minimum average values and their averages:")
print(max_min_values[['min_hour', 'min_value']])

# Plotting the histogram of average values for each hour
fig, axes = plt.subplots(3, 1, figsize=(12, 18), sharex=True)

# Plotting total_onshift_dashers histogram
axes[0].bar(hourly_averages.index, hourly_averages['total_onshift_dashers'], color='skyblue')
axes[0].set_title('Histogram of Average Total Onshift Dashers by Hour')
axes[0].set_ylabel('Average Total Onshift Dashers')
axes[0].grid(True, linestyle='--', alpha=0.7)

# Plotting total_busy_dashers histogram
axes[1].bar(hourly_averages.index, hourly_averages['total_busy_dashers'], color='lightgreen')
axes[1].set_title('Histogram of Average Total Busy Dashers by Hour')
axes[1].set_ylabel('Average Total Busy Dashers')
axes[1].grid(True, linestyle='--', alpha=0.7)

# Plotting total_outstanding_orders histogram
axes[2].bar(hourly_averages.index, hourly_averages['total_outstanding_orders'], color='salmon')
axes[2].set_title('Histogram of Average Total Outstanding Orders by Hour')
axes[2].set_ylabel('Average Total Outstanding Orders')
axes[2].set_xlabel('Hour of the Day')
axes[2].grid(True, linestyle='--', alpha=0.7)

# Set x-axis ticks for every hour
hours = hourly_averages.index
axes[2].set_xticks(hours)

plt.tight_layout()
plt.show()
```

```python
# Ensure 'created_at' is a datetime object
df['created_at'] = pd.to_datetime(df['created_at'])

# Group by 'market_id' and calculate the mean for each feature
market_averages = df.groupby('market_id').agg({
    'total_onshift_dashers': 'mean',
    'total_busy_dashers': 'mean',
    'total_outstanding_orders': 'mean'
}).reset_index()

# Define the number of plots and create the figure with a single row
num_features = len(['total_onshift_dashers', 'total_busy_dashers', 'total_outstanding_orders'])
fig, axes = plt.subplots(1, num_features, figsize=(18, 6), sharey=True)

# List of features to plot
features = ['total_onshift_dashers', 'total_busy_dashers', 'total_outstanding_orders']

# Plot each feature
for i, feature in enumerate(features):
    ax = axes[i]
    ax.bar(market_averages['market_id'], market_averages[feature])
    ax.set_title(f'Average {feature.replace("_", " ").title()} by Market ID')
    ax.set_xlabel('Market ID')
    ax.set_ylabel('Average Value')
    ax.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df_v1 = df.copy()
df_v1['store_primary_category'] = label_encoder.fit_transform(df_v1['store_primary_category'])
```

```python
df_v1.drop(['created_at', 'actual_delivery_time', 'hour'], axis=1, inplace=True)
df_v1.head()
```

```python
selected_columns = [
    'total_items', 'subtotal', 'num_distinct_items', 'min_item_price', 'max_item_price',
    'total_onshift_dashers', 'total_busy_dashers', 'total_outstanding_orders',
    'estimated_order_place_duration', 'estimated_store_to_consumer_driving_duration', 'delivery_duration_seconds'
]

# Ensure all selected columns exist in the DataFrame
selected_columns = [col for col in selected_columns if col in df_v1.columns]

# Compute the correlation matrix
corr_matrix = df_v1[selected_columns].corr()
```

```python
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='order_protocol', y='estimated_order_place_duration', data=df_v1)
plt.title('Box Plot of Order Protocol vs Estimated Order Place Duration')
plt.xlabel('Order Protocol')
plt.ylabel('Estimated Order Place Duration (seconds)')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='order_protocol', y='subtotal', data=df_v1)
plt.title('Box Plot of Order Protocol vs Subtotal')
plt.xlabel('Order Protocol')
plt.ylabel('Subtotal')
plt.show()
```

```python
# Plot the box plot for total_onshift_dashers vs market_id
plt.figure(figsize=(10, 6))
sns.boxplot(x='market_id', y='total_onshift_dashers', data=df)
plt.title('Box Plot of Market ID vs Total Onshift Dashers')
plt.xlabel('Market ID')
plt.ylabel('Total Onshift Dashers')
plt.show()

# Plot the box plot for estimated_store_to_consumer_driving_duration vs market_id
plt.figure(figsize=(10, 6))
sns.boxplot(x='market_id', y='estimated_store_to_consumer_driving_duration', data=df)
plt.title('Box Plot of Market ID vs Estimated Store to Consumer Driving Duration')
plt.xlabel('Market ID')
plt.ylabel('Estimated Store to Consumer Driving Duration (seconds)')
plt.show()

# Plot the box plot for delivery_duration vs market_id
plt.figure(figsize=(10, 6))
sns.boxplot(x='market_id', y='delivery_duration_seconds', data=df)
plt.title('Box Plot of Market ID vs Delivery Duration')
plt.xlabel('Market ID')
plt.ylabel('Delivery Duration (seconds)')
plt.show()
```

```python
avg_subtotal = df.groupby('store_primary_category')['subtotal'].mean().reset_index()
avg_subtotal = avg_subtotal.sort_values(by='subtotal', ascending=False)
```

```python
plt.figure(figsize=(10, 15))
sns.barplot(x='subtotal', y='store_primary_category', data=avg_subtotal, palette='viridis')
plt.title('Average Subtotal for Each Store Primary Category')
plt.xlabel('Average Subtotal (cents)')
plt.ylabel('Store Primary Category')
plt.show()
```

```python
df['subtotal_per_item'] = df['subtotal'] / df['total_items']
avg_subtotal_per_item = df.groupby('store_primary_category')['subtotal_per_item'].mean().reset_index()
avg_subtotal_per_item = avg_subtotal_per_item.sort_values(by='subtotal_per_item', ascending=False)
```

```python
plt.figure(figsize=(10, 15))
sns.barplot(x='subtotal_per_item', y='store_primary_category', data=avg_subtotal_per_item, palette='viridis')
plt.title('Average Subtotal per Item for Each Store Primary Category')
plt.xlabel('Average Subtotal per Item (cents)')
plt.ylabel('Store Primary Category')
plt.show()
```
